### Construction of Inverted Index.

```python
import nltk
from nltk.corpus import stopwords,PlaintextCorpusReader
from nltk.tokenize import word_tokenize, RegexpTokenizer
from collections import defaultdict

# Define Stopwords
stop_words = set(stopwords.words())

# Input: Number of Documents and Document Names
n = int(input('Enter Total Number of Documents : '))

names = []                         # File Names
vocab = []                         # Vocabulary
vocab=defaultdict(list)

# Creating Vocabulary 
for t in range(n): 
    name = "Dataset/file"+str(t+1)+".txt"
    names.append(name)
    vcb = []                       # Temporary Vocabulary for each File
    with open(name) as fin:
        file = word_tokenize(fin.read())
    for i in file:
        # Preprocessing of Data
        token = word_tokenize(i.lower())
        token = [word.lower() for word in token if word not in stop_words]
        token = [word.lower() for word in token if word.isalpha()]
        # Adding words to Temporary Vocabulary
        for item in token:
            vcb.append(item)
    # Adding words to Final Vocabulary
    for word in vcb:
        if(word in vocab.keys()):
            if (t not in vocab[word]):
                vocab[word].append(t)
        else:
            vocab[word].append(t)
    del(vcb)
    
# Creating Inverted Index
for word in vocab.keys():
    text=str(word)
    for doc_id in vocab[word]:
        text+="-->"+str(doc_id)
    print(text,file=open("invertedindex.txt","a"),end='\n')

# Output: Inverted Index
print("\nThe Inverted Index : \n")
f = open("invertedindex.txt", "r")
print(f.read()) 
```
